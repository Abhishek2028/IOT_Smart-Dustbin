```python
def getDistance(cor):
    dist = 0
    original_point = origin = (18.293346,73.55157)
    for i in range(0,len(cor)-1):
        destination = cor[i]
        cur_dist = geopy.distance.distance(origin,destination).km
        dist += cur_dist
    dist += geopy.distance.distance(origin,original_point).km

    return dist
```
